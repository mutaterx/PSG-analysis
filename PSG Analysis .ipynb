```python
import pandas as pd 
    
pd.pandas.set_option('display.max_columns', None)
data_1 = pd.read_csv("downloads/players_22.csv")

# Choosing only columns related to analysis
data_2 = data_1[[
    'nationality_name','short_name', 'club_position', 'age', 'dob', 'height_cm', 'weight_kg', 'preferred_foot', 'pace', 'movement_sprint_speed', 'movement_reactions',
    'skill_long_passing','passing', 'dribbling','movement_agility', 'skill_dribbling',
    'shooting', 'attacking_finishing', 'attacking_heading_accuracy', 'attacking_short_passing', 'skill_fk_accuracy', 'skill_ball_control', 'skill_long_passing', 
    'defending', 'defending_standing_tackle', 'defending_sliding_tackle', 
    'goalkeeping_diving', 'goalkeeping_handling', 'goalkeeping_kicking', 'goalkeeping_positioning', 'goalkeeping_reflexes', 'goalkeeping_speed',
    'power_jumping', 'power_stamina', 'power_strength', 'power_long_shots', 'power_shot_power']]

#data_3 = data_2[data_2['short_name'].isin(list_of_players_2022)]
#data_3

list_of_players_2022 = ['K. Navas', 'Sergio Rico', 'A. Letellier', 'G. Donnarumma',
    'T. Pembélé', 'A. Hakimi', 'P. Kimpembe', 'Sergio Ramos', 'Marquinhos', 'N. Mukiele',
    'M. Verratti', 'Neymar Jr', 'Juan Bernat', 'Vitinha', 'Renato Sanches', 'Nuno Mendes', 'Carlos Soler', 'M. Kapo',
    'I. Gharbi', 'K. Mbappé', 'L. Messi', 'M. Güçlü', 'Ilyes Housni', 'H. Ekitike', 
    'K. Nagera']

#Missing players
#'Danilo', Marcin Bulka, El Chadaille Bitshiabu, 'Fabian Ruiz, Warren Zaïre-Emery, Ayman Kari, Ilyes Housni, mauro icardi


pd.pandas.set_option('display.max_columns', None)
pd.pandas.set_option('display.max_rows', None)

data_3 = data_2[data_2['short_name'].isin(list_of_players_2022)]
data_3 = data_3.sort_values("short_name", ascending=True) 

#Dropping duplicates
data_3.drop(data_3[(data_3['short_name'] == 'Vitinha') & (data_3['height_cm'] == 178)].index, inplace=True)

df = data_3.reset_index()

df.head(30)
```

# Attackers Analysis 

```python
attack_name = ['Neymar Jr', 'K. Mbappé', 'L. Messi', 'H. Ekitike']
#Missing Housni Ilyes

attack = df[['nationality_name','short_name', 'age','height_cm', 'weight_kg', 'preferred_foot',
            'pace', 'movement_sprint_speed', 'movement_reactions',
            'skill_long_passing',
            'passing', 'dribbling', 'movement_agility', 'skill_dribbling',
            'shooting','power_shot_power',
            'attacking_finishing', 'attacking_heading_accuracy', 'attacking_short_passing',
            'skill_fk_accuracy', 'skill_ball_control']]

#attack finishing = scoring 


attack = attack[attack['short_name'].isin(attack_name)]
attack.head(4)
```

```python
import pandas as pd
  
# Load data into DataFrame
attack_scatter = pd.DataFrame(data = attack)
  
# Draw a scatter plot and here size of dots determined by age of person
attack_scatter.plot.scatter(x = 'short_name', y = 'movement_reactions', c = 'red')

```

# Dribbling

From the scatter plots below it is evident that the  variables 'movement_agility', 'skill_ball_control' and 'movement_reactions' corresponds to the second varibale 'dribbling', the variables have a positive correlation. 
<br>
It is thus possible to assume a relationship between the first 3 variables and the variable dribbling.

<br> 
Conclusion
<br>
We can conclude that in football the higher a player's agility, reactions, and ball control the quicker, more precise, and better they will be at dribbling. 

```python
dribbling = attack_scatter.plot.scatter(x = 'movement_agility', y = 'dribbling')
dribbling

import matplotlib.pyplot
from matplotlib import pyplot as plt
plt.title("Movement Agility")
plt.show()
```
